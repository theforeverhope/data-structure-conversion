```python
%matplotlib inline

import os
import json
import pylab as pl
import matplotlib.pyplot as plt
import PIL.Image as Image


def build_groups(list_data, file_name, img_width, img_height):
    root = {
        'direction': 'Y',
        'empty': False,
        'children': []
    }
    for list_item in list_data:
        children = list_item['childrens']
        build_child_group(children, 'X', root['children'])
    vis_list_item(root, file_name, img_width, img_height)
    return root


def build_child_group(boxes, axis, parent):
    node = {
        'direction': 'Y' if axis == 'Y' else 'X',
        'empty': False,
        'children': []
    }
    parent.append(node)

    if len(boxes) > 1:
        min_x = min(boxes, key=lambda child: child['det'][0])['det'][0]
        min_y = min(boxes, key=lambda child: child['det'][1])['det'][1]
        max_x = max(boxes, key=lambda child: child['det'][2])['det'][2]
        max_y = max(boxes, key=lambda child: child['det'][3])['det'][3]

        node['det'] = [min_x, min_y, max_x, max_y]
        node['width'] = max_x - min_x
        node['height'] = max_y - min_y

        ruler = build_projection_ruler(boxes, axis, node)
        for sec in ruler:
            if axis == 'Y':
                build_child_group(sec, 'X', node['children'])
            else:
                build_child_group(sec, 'Y', node['children'])
    else:
        if len(boxes) == 0:
            node['empty'] = True
        elif len(boxes) == 1:
            min_x = min(boxes, key=lambda child: child['det'][0])['det'][0]
            min_y = min(boxes, key=lambda child: child['det'][1])['det'][1]
            max_x = max(boxes, key=lambda child: child['det'][2])['det'][2]
            max_y = max(boxes, key=lambda child: child['det'][3])['det'][3]

            node['det'] = [min_x, min_y, max_x, max_y]
            node['width'] = max_x - min_x
            node['height'] = max_y - min_y
            node['children'].append(boxes[0])

            calculate_alignments(node)


def build_projection_ruler(boxes, axis, node):
    ruler = []
    slices = build_slices(axis, node)
    fill_slice(boxes, slices, axis)

    s = []
    is_empty = False
    ruler.append(s)

    for i in range(len(slices)):
        if 'boxes' in slices[i]:
            boxes = slices[i]['boxes']
        else:
            boxes = []

        if not is_empty:
            if slices[i]['empty']:
                s = [box for box in boxes]
                ruler.append(s)
                is_empty = True
            else:
                add_boxes(boxes, s)
        else:
            if not slices[i]['empty']:
                s = [box for box in boxes]
                ruler.append(s)
                is_empty = False
            else:
                add_boxes(boxes, s)
    return ruler


def build_slices(axis, node):
    slices = []

    precision = 1000

    if axis == 'Y':
        min_det = node['det'][1]
        max_det = node['det'][3]
    else:
        min_det = node['det'][0]
        max_det = node['det'][2]

    slice_size = (max_det - min_det) / precision

    for i in pl.frange(min_det, max_det, slice_size):
        a_slice = {
            'start': i,
            'end': i + slice_size,
            'empty': True
        }
        slices.append(a_slice)

    return slices


def fill_slice(boxes, slices, axis):
    for box in boxes:

        temp = [i for i, p in enumerate(slices) if p['start'] <= box['det'][1] <= p['end']]
        if len(temp) == 0:
            break

        if axis == 'Y':
            first = [i for i, p in enumerate(slices) if p['start'] <= box['det'][1] <= p['end']][0]
            last = [i for i, p in enumerate(slices) if p['end'] >= box['det'][3]][0]
        else:
            first = [i for i, p in enumerate(slices) if p['start'] <= box['det'][0] <= p['end']][0]
            last = [i for i, p in enumerate(slices) if p['end'] >= box['det'][2]][0]

        for i in range(first, last):
            if 'boxes' not in slices[i]:
                slices[i]['boxes'] = []
            slices[i]['boxes'].append(box)
            slices[i]['empty'] = False


def calculate_alignments(node):
    if node['direction'] == 'X' and len(node['children']) == 1:
        box = node['children'][0]
        center_b = (box['width'] / 2) + box['det'][0]
        center_g_steps = box['width'] / 3

        if center_b < center_g_steps:
            box['alignment'] = 'left'
        else:
            if center_g_steps < center_b < center_g_steps * 2:
                box['alignment'] = 'center'
            else:
                box['alignment'] = 'right'


def add_boxes(boxes, section):
    if len(section) == 0:
        for box in boxes:
            section.append(box)
    else:
        for box in boxes:
            have = False
            for sec in section:
                diff_sum = 0
                for i in range(4):
                    diff_sum += abs(sec['det'][i] - box['det'][i])
                if diff_sum == 0:
                    have = True
            if not have:
                section.append(box)
    return section


# ------------- visualize result -------------


def vis_list_item(list_item, file_name, width, height):
    fig, ax = plt.subplots(figsize=(10, 8))
    im = Image.open('./json/' + file_name + '.png')  # background image

    im = im.resize((width, height))
    ax.imshow(im, aspect='equal')

    draw_children(ax, list_item['children'], 0)

    # plt.axis('off')
    plt.show()


def draw_children(ax, comp_list, color_index):
    colors = ['royalblue', 'firebrick', 'maroon', 'red', 'grey', 'orange', 'teal', 'darkviolet']

    if len(comp_list) > 0:
        comp_list_len = len(comp_list)
        for comp in comp_list:
            if 'det' in comp:
                draw_box(ax, comp['det'], fill=False, edge_color=colors[color_index], line_width=1)

            if 'children' in comp:
                draw_children(ax, comp['children'], color_index + 1)


def draw_box(ax, rec, fill=False, edge_color='black', line_width=0.5):
    ax.add_patch(plt.Rectangle((rec[0], rec[1]), rec[2] - rec[0], rec[3] - rec[1], fill=fill, edgecolor=edge_color,
                               linewidth=line_width))


# ------------- file operation -------------


def get_files(path, file_type='', append_path=True):
    """
    get the file path list of the giving file dir path
    :param path: the path of file dir
    :param file_type
    :param append_path: if return the whole path or not
    :return: file list
    """

    all_files = os.listdir(path)

    if not file_type:
        result = all_files
    else:
        new_file_types = [file_type.lower(), file_type.upper()]
        result = filter(lambda file_name: file_name.split(".")[-1] in new_file_types, all_files)
    if append_path:
        result = [os.path.join(path, f) for f in result]
    return result


def import_json(path):
    f = open(path)
    return json.load(f)


# ------------- program begin -------------


all_json = get_files('./json')
for json_name in all_json:
    if json_name.endswith('json'):
        print(json_name, '--------------------')
        data = import_json(json_name)
        for component in data['components']:
            if component['class_name'] == 'list':
                build_groups(component['childrens'], json_name.split('/')[-1].split('.')[0], data['width'], data['height'])
```
